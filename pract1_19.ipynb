```python
df = pd.read_csv("C:/Users/gadak/OneDrive/Documents/dataset_pizza_data.csv")
df
```

```python
# Encoding categorical features
label_encoders = {}
for column in df.select_dtypes(include=['object']).columns:
    label_encoders[column] = LabelEncoder()
    df [column] = label_encoders[column].fit_transform(df [column])
```

```python
# Extract features and target variable
X = df.drop('ranking', axis=1)
y =df ['ranking']
```

```python
# Add a constant to the model (intercept)
X = sm.add_constant(X)
```

```python
# Fit an OLS (Ordinary Least Squares) model
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
#Adaptive Conjoint Analysis (ACA)
# For simplicity, we'll use PCA as a proxy for ACA in this example
pca = PCA(n_components=2)
pca_fit = pca.fit_transform(X)

plt.figure(figsize=(12, 8))
plt.scatter(pca_fit[:, 0], pca_fit[:, 1], c=y, cmap='viridis')
plt.title('Adaptive Conjoint Analysis (ACA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Ranking')
plt.show()
```

```python
# Fit a Logistic Regression model for CBC Analysis
cbc_model = LogisticRegression(max_iter=1000)
cbc_model.fit(X, y)

# Get the importance scores (coefficients)
importance_scores = cbc_model.coef_[0]
features = X.columns

# Plotting importance scores
plt.figure(figsize=(12, 8))
sns.barplot(x=features, y=importance_scores)
plt.title('Discrete Choice-Based Conjoint (CBC) Analysis')
plt.xlabel('Features')
plt.ylabel('Importance Scores')
plt.xticks(rotation=45)
plt.show()
```
